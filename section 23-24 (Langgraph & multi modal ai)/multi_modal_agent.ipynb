# Multi-modal agent

- multi-modal agent is an agent that can process and understand multiple types of data, such as text, images, audio, and video. This allows the agent to have a more comprehensive understanding of the world and to perform a wider range of tasks.

```python
from openai import OpenAI
from dotenv import load_dotenv
import os 
load_dotenv('./.env')

client = OpenAI(
    base_url="https://models.github.ai/inference",
    api_key=os.getenv("GITHUB_TOKEN"),
)
 
```

### passing an url

```python
response=client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "user",
            "content":[
                {"type":"text","text":"Generate a caption for this image in about 50 words."},
                {"type":"image_url","image_url":{"url":"https://images.pexels.com/photos/7775640/pexels-photo-7775640.jpeg"}}
            ]
        }
    ]
)

print(response.choices[0].message.content)
```
