## Method Resolution Order (MRO)


```python
class A:
    label='A : Base Class'

class B(A):
    label='B : Boom Blend'

class C(A):
    label='C: Chase Blend'

class D(C,B):
    pass 

cup=D()
print(cup.label)
print(D.__mro__)
print(A.label)
print(type(cup))
print(D.mro()[-2].label)
```

### ClassMethod Vs StaticMethod

```python
class chaiorder:
    def __init__(self,tea_type,sweetness,size):
        self.tea_type=tea_type
        self.sweetness=sweetness
        self.size=size

    @classmethod
    def from_dict(cls,order_data):
        return cls(
            order_data['tea_type'],
            order_data['sweetness'],
            order_data['size'],
        )
    
    @classmethod
    def from_String(cls,order_Str):
        tea_type,sweetness,size=order_Str.split('-')
        return cls(tea_type,sweetness,size)
    
order1=chaiorder.from_dict({'tea_type':'black','sweetness':'medium','size':'large'})
order2=chaiorder.from_String('green-high-small')
order3=chaiorder('herbal','low','medium')
print(order1.tea_type,order1.sweetness,order1.size)
print(order2.tea_type,order2.sweetness,order2.size)
print(order3.__dict__)
```

## Property Decorators

```python
class TeaLeaf:

    def __init__(self,age):
        self._age=age

    @property
    def age(self):
        return self._age + 2
    
    @age.setter
    def age(self,age):
        if 1<=age<=5:
            self._age=age
        else:
            raise ValueError('Age must be between 1 and 10')
        
leaf=TeaLeaf(5)
print(leaf.age)
leaf.age=6
print(leaf.age)
# leaf.age=15  # This will raise ValueError
```
