## Custom Exceptions

```python
class outofIngredientsError(Exception):
    pass

def make_chai(milk,sugar):
    if milk==0 or sugar==0:
        raise outofIngredientsError("We are out of ingredients")
    print("chai is ready")

make_chai(0,1)
```

## Mini Project With Exception handling

```python
class InvalidChaiError(Exception): pass

def bill(flavor,cups):
    menu={
        'masala':10,
        'ginger':40,
        'elaichi':15
    }
    try:
        if flavor not in menu:
            raise InvalidChaiError(f"{flavor} is not available")
        if not isinstance(cups,int):
            raise ValueError("Cups must be an integer")
        total=menu[flavor]*cups
        print(f"your bill is {total}")
    except Exception as e:
        print("Error: ",e)
    finally:
        print("Thank you for visiting chai shop")

bill('lemon',2)
bill('masala','two')
bill('ginger',3)
```
