Definition of Comprehensions: They allow you to create collections like lists, sets, dictionaries, or generators in just one line of code.

Real-life Applications: Comprehensions are useful for filtering items, transforming data, creating new collections, and flattening nested structures.
Benefits: Using comprehensions leads to cleaner code, faster execution, and reduced memory usage compared to traditional looping methods.
Types of Comprehensions: The lecture outlines four types: list comprehensions, set comprehensions, dictionary comprehensions, and generator comprehensions.

## List Comprehension

```python
menu =[
    'Home',
    'About',
    'Services',
    'Contact',
    'Blog'
]
iced_menu=[item for item in menu if len(item)>4 ]
print(iced_menu)
```

## Set Comprehensions

```python
example=[
    'apple','apple','guava',
    'banana',  'banana','kiwi',
    'mango', 'orange','kiwi'
]
print(example)
unique_example={item for item in example}
print(unique_example)
```

```python
recipes={
    'apple pie': ['apple', 'flour', 'sugar', 'butter'],
    'banana bread': ['banana', 'flour', 'sugar', 'eggs'],
    'fruit salad': ['apple', 'banana', 'orange', 'kiwi']
    }


unique_ingredients={spice for ingredients in recipes.values() for spice in ingredients}
print(unique_ingredients)
```

## Dictionary Comprehension

```python
car_prices_inr={
    'Toyota': 200,
    'Honda': 1800,
    'Ford': 2200,
    'BMW': 5000
}
car_prices_usd={brand:price/80 for brand,price in car_prices_inr.items()}
print(car_prices_usd)
```

## Generator Comprehension
* used mainly for saving memory (lazy evaluation)
<br>
![Generator Comprehension Diagram](./comprehensio.excalidraw.png)

```python
daily_sales=[500, 700, 800, 600, 750, 900, 650]

total_sales=sum([sale for sale in daily_sales if sale>500])
print(total_sales)
```
