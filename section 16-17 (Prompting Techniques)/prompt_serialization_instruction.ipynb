## Alpaca Prompting Template
The Alpaca prompting template is a widely used format for structuring prompts when interacting with language models. It consists of three main components: Instruction, Input, and Output. The template is designed to provide clear guidance to the model on how to process the input and generate the desired output.

```python
# Prompt Styles

### Instructions : <SYSTEM_PROMPT> \n

### Input: <USER_QUERY> \n

### REsponse: \n 

response=client.chat.completions.create(
    model="openai/gpt-4o",
    response_format={"type":"json_object"},
    messages=[
        {"role":"system","content":"You are a helpful assistant that helps with coding problems."},
        {"role":"user","content":"Write a python function to check if a number is prime or not."}
        ]
)
```

## ChatML Prompting Template
The ChatML prompting template is a structured format for creating prompts that guide language models in generating responses in a conversational manner. It typically includes components such as System, User, and Assistant messages, allowing for a more interactive and dynamic exchange between the user and the model. This template is particularly useful for tasks that require multi-turn interactions or when you want to simulate a conversation with the model.

```python

# Chat ML Prompting template

response=client.chat.completions.create(
    model="openai/gpt-4o",
    response_format={"type":"json_object"},
    messages=[
        {"role":"system","content":"You are a helpful assistant that helps with coding problems."},
        {"role":"user","content":"Write a python function to check if a number is prime or not."}
        ]
)
```

## INST Prompting Template
- used by llama 3 and llama 4
The INST prompting template is a format that focuses on providing clear instructions to the language model. It typically includes an Instruction component that outlines the task or question, and may also include an Input component for any necessary context or data. The Output component is where the model's response is expected. This template is designed to be straightforward and is often used for tasks that require specific instructions or when the desired output is well-defined.

```python
[INST] what is the capital of france? [/INST]
```
